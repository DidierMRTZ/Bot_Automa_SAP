```python
dic={}
for i in range(0,100):
    r=random.randrange(0,len(letras)-1)
    r2=random.randrange(0,len(letras)-1)
    dic["people"+str(i)]=letras[r]+letras[r2]
```

```python
test=pd.DataFrame(pd.Series(dic)).reset_index()
```

```python
dic_Na={}
def recur(dataframe,start,end):
    for i in range(start,end):
        if test[0][i]=="aa":
            person=test["index"][i]
            valor=test[0][i]
            dic_Na[person]=valor
            print(test["index"][i]+"-----"+test[0][i])
        else:
            try:
                recur(dataframe,i+1,i+15)
                break
            except:
                break
```
